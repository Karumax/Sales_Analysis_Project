```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 
import seaborn as sns
```

```python
df=pd.read_csv('Diwali Sales Data.csv', encoding='unicode_escape')
```

```python
df.head(10)
```

```python
df.info()
```

```python
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
pd.isnull(df).sum()
```

```python
df['Amount']=df['Amount'].astype('int')
```

```python
df['Amount'].dtype
```

```python
df.columns
```

```python
df.rename(columns={'Marital_Status':'Shaadi_shuda'}).head(10)
```

```python
df.describe()
```

# Sales

```python
df[['Age','Orders','Amount']].describe()
```

```python
sns.countplot(x = 'Gender',data = df)
```

```python
ax=sns.countplot(x = 'Gender',data = df)

for bars in ax.containers:
     ax.bar_label(bars)
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_gender=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=sales_gender)
```

# Age

```python
df.columns
```

```python
sns.countplot(data=df,hue='Gender',x='Age Group')
```

```python
ax=sns.countplot(data=df,hue='Gender',x='Age Group')

for bar in ax.containers:
    ax.bar_label(bar)
```

```python
Sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=Sales_age) 
```

# State

```python
df.columns
```

```python
state_sales=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(data=state_sales,x='State',y='Orders')
```

```python
state_sales=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(data=state_sales,x='State',y='Amount')
```

# Maratial Status

```python
ax=sns.countplot(data=df,x='Marital_Status')

sns.set(rc={'figure.figsize':(3,4)})
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
state_sales=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(4,5)})
sns.barplot(data=state_sales,x='Marital_Status',y='Amount',hue='Gender')
```

# Occupation

```python
ax=sns.countplot(data=df,x='Occupation')

sns.set(rc={'figure.figsize':(20,5)})
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
state_sales=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=state_sales,x='Occupation',y='Amount')
```

# Product Category

```python
ax=sns.countplot(data=df,x='Product_Category')

sns.set(rc={'figure.figsize':(20,5)})
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
state_sales=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=state_sales,x='Product_Category',y='Amount')
```

```python
state_sales=df.groupby(['Product_ID'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=state_sales,x='Product_ID',y='Amount')
```
